```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import time

# Create the figure and axes
fig, ax = plt.subplots()

# Initialize an empty line object (we will update this later)
line, = ax.plot([], [])

# Set the x and y axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

# Function to update the plot at each animation frame
def update(frame):
    x = np.linspace(0, 10, 1000)  # Generate x values
    y = np.sin(2 * np.pi * (x - 0.01 * frame))  # Generate y values (a sine wave)
    line.set_data(x, y)  # Update the line data
    return line,

# Create the animation
animation = FuncAnimation(fig, update, frames=100, interval=50, blit=True)

# Display the animation
# plt.close(animation._fig)  # Prevents displaying the static plot below the animation
# animation
```
